```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = "C:/Users/sweet/Downloads/trending.csv/Trending Movies Dataset.csv"
data = pd.read_csv(file_path)

# Step 1: Data Cleaning
# Drop the irrelevant column
data = data.drop(columns=['Unnamed: 0'])

# Handle missing values
data['original_title'].fillna('Unknown', inplace=True)
data['release_date'].fillna('Unknown', inplace=True)

# Step 2: Explore the Data's Characteristics( Summary Statistics)
print("Dataset Shape:", data.shape)
print("\nDataset Info:")
print(data.info())
print("\nStatistical Summary:")
print(data.describe())

# Step 3:Distributions(Visualizing -Histogram for numerical columns)
numerical_cols = ['popularity', 'vote_average', 'vote_count']
data[numerical_cols].hist(bins=30, figsize=(12, 8), color='skyblue')
plt.suptitle('Histograms of Numerical Columns', fontsize=16)
plt.show()

#Correlations: Visualizations: Heatmap and scatter plots.
#Examines relationships between numerical variables and highlights strong correlations.
# Step 4: Correlation Heatmap
# Calculate the correlation matrix
correlation_matrix = data[numerical_cols].corr()

# Plot the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()

# Step 5: Scatter Plots (Relationships between variables)
plt.figure(figsize=(12, 6))
sns.scatterplot(data=data, x='popularity', y='vote_count', hue='media_type')
plt.title('Scatter Plot: Popularity vs. Vote Count')
plt.show()

# Step 6: Outliers (Visualizations: Boxplots.)
plt.figure(figsize=(12, 6))
sns.boxplot(data=data, x='media_type', y='popularity', palette='Set2')
plt.title('Boxplot: Popularity by Media Type')
plt.show()

# Optional: Distribution of Categorical Variables
plt.figure(figsize=(8, 5))
sns.countplot(data=data, x='media_type', palette='pastel')
plt.title('Count of Media Types')
plt.show()

# Step 7: Save Cleaned Dataset (Optional)
cleaned_file_path = "C:/Users/sweet/Downloads/Trending Movies Dataset_cleaned.csv"
data.to_csv(cleaned_file_path, index=False)
print(f"Cleaned data saved to {cleaned_file_path}")
```
